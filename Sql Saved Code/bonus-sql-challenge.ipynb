```python
# Pandas
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats

# # SQL Alchemy
# from sqlalchemy import create_engine
# # engine = create_engine('postgresql://localhost/sql-challenge')
# # engine = create_engine('postgresql://localhost:5432/sql-challenge')
# # from config import Password
# # engine = create_engine('postgresql://postgres:Password@localhost/sql-challenge')
# conn_admin = psycopg2.connect("dbname=sql-challenge user=username host=localhost")
# conn = engine.connect()

salaries_path = "Data/salaries.csv"
employees_path = "Data/employees.csv"
titles_path = "Data/titles.csv"
```

```python
salaries = pd.read_csv(salaries_path)
employees = pd.read_csv(employees_path)
titles = pd.read_csv(titles_path)
```

```python
employee_df = pd.merge(salaries, employees, how="left", on=["emp_no", "emp_no"])
employee_df
```

```python
# Create a histogram to visualize the most common salary ranges for employees.

plt.hist(salaries['salary'], 20, density=False, label="Salary", color='purple')
plt.ylabel("Employees")
plt.legend() 
plt.title("Salaries")
plt.savefig("Common Salary Ranges.png")
```

```python
# Merging
salary_df = pd.merge(employee_df, titles, left_on='emp_title_id', right_on='title_id')
salary_df
```

```python
# Create a bar chart of average salary by title
avg_salaries_df = salary_df.groupby(['title'])['salary'].mean()
avg_salaries_df.sort_values(ascending=False).plot.bar(color='violet', align='center')
plt.title("Average Salaries By Title")
plt.ylabel("Salary")
plt.xlabel("Title")
plt.show()
```
